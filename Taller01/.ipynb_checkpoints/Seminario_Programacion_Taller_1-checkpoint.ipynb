## **2.2. MANEJO DE DATOS ATIPICOS**

Para realizar un adecuado manejo de datos atipicos en nuestra base de datos, vamos a generar en primera medida un pequeño resumen estadistico de la variable valor de contrato, donde podemos observar que el rango de nuestros datos es bastante amplio. Sin embargo el percentil 75 se situa en un valor de contrato de 39 millones:

```python
stats = pd.DataFrame(secop["valor_del_contrato"].describe()).style.format("{:,.0f}")
stats
```

Para evidenciar lo anterior, podemos generar un boxplot para ilustrar la presencia de datos atipicos en la variable valor de contrato:

```python
sns.boxplot(x=secop["valor_del_contrato"])
```

Dado lo anterior, realizamos definimos nuestros valores atipicos como aquellos valores que se encuentran a 1.5 veces el rango intercuartilico de nuestra variable y procedemos a remplazarlos por "nan":

```python
for x in ["valor_del_contrato"]:

    q75,q25 = np.percentile(secop.loc[:,x],[75,25])

    intr_qr = q75-q25
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
    
    secop.loc[secop[x] < min,x] = np.nan
    secop.loc[secop[x] > max,x] = np.nan
```

Luego de lo anterior, estiamos el porcentaje de datos atipicos en nuestra variable encontrando que corresponde a un 9.98%. Que para nuestro criterio no es considerado como un porcentaje significativo de datos perdidos, por lo cual procedemos a realizar una imputación a traves de la mediana de los datos:

```python
secop["valor_del_contrato"].isnull().sum()/len(secop)*100
```

```python
for i in ["valor_del_contrato"]:
    secop.loc[secop.loc[:,i].isnull(),i]=secop.loc[:,i].median()
```

Verificamos que la imputación:

```python
secop["valor_del_contrato"].isnull().sum()/len(secop)
```

Relizamos nuevamente un resumen estadistico de nuestra variable, observando que el valor maximo ya no se encuenta tan alejado del percentil 75

```python
stats = pd.DataFrame(secop["valor_del_contrato"].describe()).style.format("{:,.0f}")
stats
```

Realizamos un boxplot de nuestra variable para corroborar nuestra imputación para manejar los datos atipicos presentes originalment en nuestra base:

```python
sns.boxplot(x=secop["valor_del_contrato"])
```

Luego de darle tratamiento a los datos atipicos de la variable valor del contrato podemos relacionarla con las demas variables que son de nuestro interes y generar los correspondientes boxplots:

```python
def grafi_box(i):
        
        dfg = secop.groupby(i)["valor_del_contrato"].sum()\
               .sort_values(ascending = False)
        
        fig, axs = plt.subplots(figsize=(20,16))
        axs=sns.boxplot(data=secop,y='valor_del_contrato',x=i)
        axs.set_xticklabels(axs.get_xticklabels(), rotation=90);
        
        plt.show()
    
interact(grafi_box,\
        i=widgets.Dropdown(options=filtro_colum,\
                           value='departamento',\
                           description='Variable: '))  
```

En estos términos, vemos que el valor per capita de los contratos aparecen jurisdicciones en el en niveles similares al de Bogotá. Este análisis nos perite obervar una desviación relevante: el valor promedio de los contratos sin departamente definido es 20% superior a valor de la ciudad con el mayor valor promedio, los cuales representan un total de $29 mil millones.

```python
secop2 = pd.DataFrame(secop.groupby("departamento")["valor_del_contrato"].sum().map('{:,.0f}'.format))
secop2["cantidad"]=secop.departamento.value_counts()
secop2["percapita"]=(secop.groupby("departamento")["valor_del_contrato"].sum()/secop.departamento.value_counts()).map('{:,.0f}'.format)
secop2.reset_index()
secop2.sort_values(by="percapita", ascending=False)

```
